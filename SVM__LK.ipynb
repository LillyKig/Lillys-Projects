```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

from matplotlib import pyplot as plt
%matplotlib inline
```

```python

df = pd.read_csv ("/content/data.csv") # Dataset - Breast Cancer Wisconsin Data
df['diagnosis'] = df['diagnosis'].map({'M': 1,'B': 2})
# Label values - 1 for Malignant and 2 for Benign
labels = df['diagnosis'].tolist() #creating new column
df['Class'] = labels  #Copying values of diagnosis to newly created labels column
df = df.drop(['id', 'Unnamed: 32', 'diagnosis'], axis=1)  #Dropping unncessary columns
df.head(5)  #Displaying first five rows of the dataset
```

```python
target_names = ['', 'M', 'B']
df['attack_type'] = df.Class.apply(lambda x: target_names[x])
df.head(5)
```

```python
df1 = df[df.Class == 1] #dividing the classes
df2 = df[df.Class == 2]
```

```python
plt.xlabel('radius_mean')
plt.ylabel('symmetry_mean')
plt.scatter(df1['radius_mean'], df1['symmetry_mean'], color='green', marker='+')
plt.scatter(df2['radius_mean'], df2['symmetry_mean'], color='blue', marker='.')
```

```python
X = df.drop(['Class', 'attack_type'], axis='columns')
X.head()
```

```python
y = df.Class
y.head(10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
print(len(X_train))
print(len(X_test))
```

```python
model = SVC(kernel='linear')
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
print(predictions)
```

```python
percentage = model.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
res = confusion_matrix(y_test, predictions)
print("Confusion Matrix")
print(res)
print(f"Test Set: {len(X_test)}")
print(f"Accuracy = {percentage*100} %")
```
