**Frequency distribution of the target variable (Species) on the selected flowers**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris['Species'].value_counts()
```

```python
sns.pairplot(iris, hue="Species")
```

Findings:

The largest species of Iris is Iris-setosa, followed by Iris-versicolor and the smallest is Iris-virginica.

```python
numeric_features = iris.columns[:-1]
iris[numeric_features].hist(figsize=(12, 8), bins=20)
plt.suptitle('Histograms for Numeric Features in Iris Dataset', y=1.02)
plt.show()
```

```python
# Create histograms for each feature grouped by species
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i+1)
    sns.histplot(iris, x=feature, hue='Species', kde=True, element='step', stat='density', common_norm=False)
    plt.title(f'Histogram of {feature} by Species')

plt.tight_layout()
plt.show()
```

```python
# Create bar charts for each feature
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i+1)
    sns.barplot(x='Species', y=feature, data=iris)
    plt.title(f'Bar Chart of {feature} by Species')

plt.tight_layout()
plt.show()
```
