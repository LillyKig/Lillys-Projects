```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the Excel file
file_path = "Downloads/All_DEWA_Bills_2023_2025.xlsx"
df = pd.read_excel(file_path)

# Preview the data
df
```

## Data Cleaning

```python
# Convert date columns to datetime

df["Period Start"] = pd.to_datetime(df["Period Start"], errors='coerce', dayfirst=True)
df["Period End"] = pd.to_datetime(df["Period End"], errors='coerce', dayfirst=True)

# Extract year and month
df["Year"] = df["Source File"].dt.year
df["Month"] = df["Source File"].dt.strftime('%B')
df["Month_Num"] = df["Source File"].dt.month

# Sort for chronological order
df = df.sort_values(["Year", "Month_Num"])
df
```

```python

print(df["Month"].unique())

```

```python
# Convert Issue Date to month abbreviation
df["Month"] = df["Period Start"].dt.strftime("%b")  # %b gives Jan, Feb, etc.
df["Month"] 
```

```python
# Define the correct order
month_order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
               "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Convert to ordered categorical
df["Month"] = pd.Categorical(df["Month"], categories=month_order, ordered=True)

print(df["Month"].unique())
df = df.sort_values(["Year", "Month"])
```

## Descriptive Analysis

```python
df[["Electricity (AED)", "Water (AED)", "Total (AED)"]].describe()

```

```python
#Total Annual Spending
df.groupby("Year")[["Electricity (AED)", "Water (AED)", "Total (AED)"]].sum()
```

```python
#Average Monthly Spending
df.groupby("Month")[["Electricity (AED)", "Water (AED)", "Total (AED)"]].mean()
```

```python
#Highest and lowest Bill

# Find the row with the highest electricity bill
highest_electricity = df.loc[df["Electricity (AED)"].idxmax()]

# Find the row with the lowest water bill
lowest_water = df.loc[df["Water (AED)"].idxmin()]

# Combine both results into one table
extreme_bills = pd.DataFrame([highest_electricity, lowest_water])
extreme_bills
```

```python
#Monthly Trends

monthly_avg = df.groupby("Month")[["Electricity (AED)", "Water (AED)"]].mean()
monthly_avg = monthly_avg.reindex(["Jan", "Feb", "Mar", "Apr", "May", "Jun",
                                   "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])

monthly_avg.plot(kind="bar", figsize=(10, 5))
plt.title("Average Monthly Utility Charges")
plt.ylabel("AED")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Seasonal Analysis
def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Autumn"

df["Season"] = df["Month_Num"].apply(get_season)
df.groupby("Season")[["Electricity (AED)", "Water (AED)"]].mean()
```

```python


# Calculate seasonal averages
seasonal_avg = df.groupby("Season")[["Electricity (AED)", "Water (AED)"]].mean().reindex(["Winter", "Spring", "Summer", "Autumn"])

# Plot the seasonal averages
seasonal_avg.plot(kind="bar", figsize=(10, 6))
plt.title("Average Utility Charges by Season")
plt.ylabel("Average Charges (AED)")
plt.xlabel("Season")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

## Visualize Monthly Electricity and Water Charges

```python
import matplotlib.pyplot as plt

# Group data by Period Start
df_plot = df.sort_values(["Year", "Month_Num"])

plt.figure(figsize=(12, 6))
plt.plot(df_plot["Period End"], df_plot["Electricity (AED)"], label="Electricity (AED)", marker='o')
plt.plot(df_plot["Period End"], df_plot["Water (AED)"], label="Water (AED)", marker='s')
plt.title("Monthly Electricity and Water Charges")
plt.xlabel("Month")
plt.ylabel("Charges (AED)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Year-over-Year Comparison for Each Month

```python
pivot_df = df.pivot_table(index=df["Month"], columns="Year", values=["Electricity (AED)", "Water (AED)"], observed=False)
pivot_df = pivot_df.sort_index()  # Uses the calendar order now

pivot_df
```

##  Calculate and Plot Year-over-Year Change

```python



# --- Plot Electricity ---
plt.figure(figsize=(10, 5))
for year in pivot_df["Electricity (AED)"].columns:
    plt.plot(pivot_df.index, pivot_df["Electricity (AED)"][year], marker='o', label=f"Electricity {year}")
plt.title("Electricity Charges Year-on-Year")
plt.xlabel("Month")
plt.ylabel("Electricity (AED)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# --- Plot Water ---
plt.figure(figsize=(10, 5))
for year in pivot_df["Water (AED)"].columns:
    plt.plot(pivot_df.index, pivot_df["Water (AED)"][year], marker='s', label=f"Water {year}")
plt.title("Water Charges Year-on-Year")
plt.xlabel("Month")
plt.ylabel("Water (AED)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Calculate Year on Year Change

```python
# Calculate YoY % change (example: 2024 vs 2023 if both exist)
pivot_df["Electricity Change (%)"] = (
    (pivot_df[("Electricity (AED)", 2024)] - pivot_df[("Electricity (AED)", 2023)]) / pivot_df[("Electricity (AED)", 2023)]
) * 100

pivot_df["Water Change (%)"] = (
    (pivot_df[("Water (AED)", 2024)] - pivot_df[("Water (AED)", 2023)]) / pivot_df[("Water (AED)", 2023)]
) * 100

pivot_df[["Electricity Change (%)", "Water Change (%)"]]
```

```python
import matplotlib.pyplot as plt

# --- Plot YoY Electricity % Change ---
plt.figure(figsize=(10, 5))
plt.bar(pivot_df.index, pivot_df["Electricity Change (%)"], color='orange')
plt.axhline(0, color='gray', linestyle='--')
plt.title("Year-over-Year % Change in Electricity Charges (2024 vs 2023)")
plt.xlabel("Month")
plt.ylabel("Change (%)")
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Plot YoY Water % Change ---
plt.figure(figsize=(10, 5))
plt.bar(pivot_df.index, pivot_df["Water Change (%)"], color='teal')
plt.axhline(0, color='gray', linestyle='--')
plt.title("Year-over-Year % Change in Water Charges (2024 vs 2023)")
plt.xlabel("Month")
plt.ylabel("Change (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Forecast 2025 using seasonal Averages

```python
# Filter past years' data
past_df = df[df["Year"].isin([2024])]

# Calculate monthly averages
monthly_avg = past_df.groupby("Month_Num")[["Electricity (AED)", "Water (AED)"]].mean()
monthly_avg["Month"] = pd.to_datetime(monthly_avg.index, format='%m').strftime('%B')

# Project 2025
forecast_2025 = monthly_avg.copy()
forecast_2025["Year"] = 2025
forecast_2025 = forecast_2025[["Month", "Electricity (AED)", "Water (AED)"]].reset_index(drop=True)
forecast_2025
```

```python
plt.figure(figsize=(10, 5))
plt.plot(forecast_2025["Month"], forecast_2025["Electricity (AED)"], marker='o', label="Forecasted Electricity")
plt.plot(forecast_2025["Month"], forecast_2025["Water (AED)"], marker='s', label="Forecasted Water")
plt.title("Forecasted Utility Charges for 2025")
plt.xlabel("Month")
plt.ylabel("Charges (AED)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
